```python
import pandas as pd

# 假设df是您已有的Pandas DataFrame

# 保留area_name为"U.S."
gasoline_stock = df[df['area-name'] == 'U.S.']

# 保留所需的列
gasoline_stock = gasoline_stock[['period', 'value', 'units']]

# 转换period列为日期格式
gasoline_stock['period'] = pd.to_datetime(gasoline_stock['period'])

# 打印结果或进行进一步处理
print(gasoline_stock)
```

```python
def get_graph(df_filtered,a):
    # 选择需要计算的列
    df_seasonal = df_filtered[['period', 'value']]

    # 提取年份和月份
    df_seasonal['year'] = df_seasonal['period'].dt.year
    df_seasonal['month'] = df_seasonal['period'].dt.month

    # 获取最近的一年和第二近的一年
    latest_year = df_seasonal['year'].max()
    second_latest_year = df_seasonal['year'].unique()[1]

    # 使用 seaborn 绘制季节性图
    plt.figure(figsize=(12, 6))

    # 设置颜色映射
    cmap = sns.color_palette("Set2", n_colors=len(df_seasonal['year'].unique()))
    # 将最近的一年设为红色
    cmap[latest_year - df_seasonal['year'].min()] = (1.0, 0.0, 0.0)

    # 将第二近的一年设为黑色
    cmap[second_latest_year - df_seasonal['year'].min()] = (0.0, 0.0, 0.0)


    # 绘制季节性图，根据年份设置颜色
    sns.lineplot(x='month', y='value', hue='year', palette=cmap, data=df_seasonal,marker='o') #

    # 设置横坐标范围
    plt.xticks(range(1, 13))

    # 设置x轴限制范围，让最左侧和最右侧紧贴数据
    plt.xlim(1, 12)

    # 设置图例
    plt.legend(title='Year', loc='upper right')

    plt.title(a)
    plt.xlabel('month')
    plt.ylabel('Stock')
    plt.grid(True)
    plt.show()
```

```python
get_graph(gasoline_stock,"US gasoline stock MMBL")
```
